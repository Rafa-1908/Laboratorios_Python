```python
from openpyxl import load_workbook
from tabulate import tabulate

# Nombre del archivo Excel
nombre_archivo = "entrada_productos_2023.xlsx"

# Cargar el archivo Excel
wb = load_workbook(filename=nombre_archivo)
sheet = wb.active

# Obtener los encabezados de las columnas
headers = []
for cell in sheet[2]:
    headers.append(cell.value)
    
#  Obtener todos los datos de la hoja activa
datos = []
for row in sheet.iter_rows(min_row=3, values_only=True):
    datos.append(row)
    
# Mostrar los datos en forma de tabla usando tabulante
tabla = tabulate(datos, headers=headers, tablefmt="pretty")

# Imprimir la tabla
print(tabla)

# Cerrar el archivo Excel después de su uso
wb.close       

    
```

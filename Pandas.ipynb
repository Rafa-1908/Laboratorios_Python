## Analizando una base de datos
https://datos.bancomundial.org/indicator/SP.POP.TOTL

```python
dfpob=pd.read_csv('datasets/API_SP.POP.TOTL_DS2_es_csv_v2_31996.csv',skiprows=3)
display(dfpob.head())
```

```python
dfpob.set_index('Country Code',inplace=True)
```

```python
dfpob.head()
```

```python
# Genera una lista de cadenas que representan los años desde 1960 hasta 2020
# La comprensión de lista (list comprehension) convierte cada número en el rango a cadena (str)
dfpob2 = dfpob[[str(k) for k in range(1960, 2021)]]

# Muestra el DataFrame 'dfpob2' que ahora contiene solo las columnas de los años entre 1960 y 2020
display(dfpob2)
```

```python
from matplotlib.figure import Figure
```

```python
# Lista de países que se quieren graficar
paises = ['MEX', 'PER', 'COL', 'ESP']  # Países seleccionados: México, Perú, Colombia, España
# Filtramos el DataFrame 'dfpob2' para que contenga solo las filas correspondientes a los países seleccionados
dfpob4 = dfpob2.loc[paises]
display(dfpob4) # Mostramos el DataFrame filtrado
# Configuramos el tamaño de la figura para la gráfica (12 unidades de ancho, 6 unidades de alto)
plt.figure(figsize=(12, 6))
for pais in paises:   # Bucle para graficar cada país
    plt.plot(dfpob4.loc[pais], label=pais)  # Graficamos la población de cada país, con 'pais' como etiqueta
plt.legend() # Añadimos una leyenda para identificar los países en la gráfica
plt.grid()  # Añadimos una cuadrícula a la gráfica
# Definimos los valores en el eje X, con un intervalo de 5 años desde 1960 hasta 2020
xvalues = np.arange(1960, 2021, 5)
# Convertimos estos valores de años (enteros) a cadenas para usarlas como etiquetas en el gráfico
xvalues = [str(k) for k in xvalues]
plt.xticks(xvalues)  # Colocamos estas etiquetas de los años en el eje X
plt.show()   # Mostramos la gráfica
```

```python
plt.figure(figsize=(12, 6))  # Configura el tamaño de la figura (ancho 12, alto 6)
ancho_barras = 0.35  # Define el ancho de las barras
plt.grid()  # Activa la cuadrícula
indice_barras = np.arange(4)  # Crea un arreglo con los índices para las barras (para 4 países)
# Grafica las barras para el año 1960
plt.bar(indice_barras, dfpob4['1960'], width=ancho_barras, label='1960')
# Grafica las barras para el año 2020 desplazando el índice por el ancho de las barras
plt.bar(indice_barras + ancho_barras, dfpob4['2020'], width=ancho_barras, label='2020')
plt.xticks(indice_barras + ancho_barras, dfpob4.index)  # Coloca los nombres de los países en el eje X
plt.legend()  # Muestra la leyenda para identificar las barras de 1960 y 2020
plt.show()  # Muestra la gráfica
```

```python
plt.figure(figsize=(12, 6))
plt.title('1960')
plt.pie(dfpob4['1960'],labels=dfpob4.index,shadow=True,autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title('2020')
plt.pie(dfpob4['2020'],labels=dfpob4.index,shadow=True,autopct='%1.1f%%')
plt.show()
```

## Combinando DataFrames

```python
General=pd.DataFrame({'nombre':['Pepito','Maria','Ana','Bob'],
                      'edades':[16,12,17,18],
                      'escolaridad':['P','S','U','U']})
display(General)

Cursos=pd.DataFrame({'nombre':['Maria','Ana','Pepito'],
                     'cursos':['Marth','Science','Data Science'],
                     'calificacion':[9,8,10],
                     'asistencias':[25,24,26]})
display(Cursos)
```

```python
pd.merge(General[['nombre','edades']],Cursos[['nombre','asistencias',]],how='outer')
```

```python
Estudiantes1=pd.DataFrame({'nombre':['Pepito','Maria','Ana'],
                          'edades':[16,12,17]})
display(Estudiantes1)
Estudiantes2=pd.DataFrame({'nombre':['Juan','Pedro','Bob'],
                          'edades':[15,14,17]})
display(Estudiantes2)
```

```python
pd.concat([Estudiantes1,Estudiantes2],ignore_index=True,axis=0) 
#Con el axis se puede concatenar de manera horizontal, eso es una alternativa al orden vertical, claro dependiendo de la situación y los datos
```

## Funciones y mapping

```python
Datos = pd.DataFrame({'Matematica': [10,8,7],   # Creación del DataFrame con tres columnas de materias
                      'Fisica': [6,5,9],       # y tres estudiantes (Pepito, Ana, María) como índice
                      'Programacion': [7,9,10]}, index=['Pepito', 'Ana', 'Maria'])
display(Datos)  # Mostrar el DataFrame completo

matematica = Datos['Matematica']  # Extrae la columna de Matemática
display(matematica)  # Mostrar los valores de la columna Matemática
```

```python
# Definir una función que reste uno al valor dado
def restauno(n):  
    return n - 1  # Retorna el valor original menos 1

# Aplicar la función 'restauno' a cada valor de la columna 'matematica'
matematica.apply(restauno)  # Usar 'apply' para aplicar la función en cada valor de la columna
```

```python
# Aplicar la función 'restauno' a todos los valores del DataFrame
Datos.applymap(restauno)  # 'applymap' aplica la función en cada celda de todo el DataFrame
```

```python
def promedio(lista):  # Define la función 'promedio' que recibe una lista de valores
    return sum(lista) / len(lista)  # Calcula el promedio de la lista dividiendo la suma entre la cantidad de elementos
# Aplica la función 'promedio' a cada fila del DataFrame 'Datos'
Datos.apply(promedio, axis=1)  # 'axis=1' indica que la función se aplicará a lo largo de las filas (no columnas)
```

```python
# Aplica las funciones 'min', 'max' y 'promedio' a cada fila del DataFrame 'Datos'
Datos.agg([min, max, promedio], axis=1)  # 'axis=1' indica que se aplicarán a las filas
```

## Tablas pivote

```python
Tabla=pd.DataFrame({
    'tipo':['moto','auto','moto','moto','auto','auto','moto','moto'],
    'velocidad':[12,23,34,26,43,21,28,30],
    'propietario':['Juan','Jorge','Maria','Ana','Juan','Maria','Bob','Camila'],
    'nuevo':['si','no','no','si','no','si','si','no'],
    'precio':[5000,6000,7500,5500,8000,4000,6500,10000]
})
display(Tabla)
```

```python
Tabla.pivot_table(index='tipo',columns='nuevo',values='precio',aggfunc='mean')
```

```python
Tabla.pivot_table(index='tipo',columns='nuevo',values='velocidad',aggfunc='min')
```

```python
Tabla.pivot_table(index='tipo',columns='nuevo',values='precio',aggfunc='count')
```

```python
Tabla.pivot_table(index='propietario',columns='tipo',values='velocidad',aggfunc='count').fillna(0)
```
